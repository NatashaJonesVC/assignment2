#### Q19- Calculate the average star rating for movies 2 hours or longer, and compare that with the average star rating for movies shorter than 2 hours.

```python
#A: I divided the data base into 'long movies' (120 minutes ie. 2 hours or longer) and 'short movies' (less than 120 minutes). I then calculated the average star rating for long and short movies, rounded the average by 2 dp and tabulated the results to be able to compare the average star rating for long, short and all movies. 
#I chose to add in an 'all movies' column to be able to compare the whole data set. From the results, we could infer that longer movies have a slightly better star rating than shorter movies.

from tabulate import tabulate
long_movies = imdb[imdb['duration'] >= 120]
short_movies = imdb[imdb['duration']<120]

lmavg = round(long_movies['star_rating'].mean(), 2)
smavg = round(short_movies['star_rating'].mean(), 2)
avg = round(imdb['star_rating'].mean(), 2)


print(tabulate([['Long Movies', lmavg], ['Short Movies', smavg], ['All Movies', avg]], headers=['Movie Length', 'Average Star Rating']))


```

```python
#Q2: How many actors have had the first credit of a movie (ie. the lead) in a movie 5 times or more? 

lead_actors = imdb['actor1'].value_counts()

df_lead = lead_actors.rename_axis('actors').reset_index(name='counts')

df_lead[df_lead["counts"]>=5]

#A: 31 actors have appeared as the lead in a movie 5 times or more
```

```python
#Visualise the number of appearances by lead actors: 

plt.figure(figsize=(200, 40))
plt.bar(df_lead['actors'], df_lead['counts'], color = 'r')
plt.ylabel("Appearances")
plt.xlabel("Actor")
plt.title("Lead Actor Appearances");


```

```python
#Q3 Who are the 10 best lead actors according to average star rating who have appeared in 5 or more movies?

#From the lead actors who have appeared in 5 or more movies, Charlie Chaplin gives the best average performance across his movies, followed by Elijah Wood, James Stedward and Jack Nicholson. Interestingly, Al Pacino leads in the highest rated movie in this list (with 9.1 star rating).
leading_leads = df_lead[df_lead["counts"]>=5]

leading_actor_movies = imdb[imdb['actor1'].isin(leading_leads['actors'])]

actor_data = leading_actor_movies[['actor1', 'star_rating']]

grouped = actor_data.groupby("actor1").agg({'star_rating': [np.min,np.max, np.mean]})

grouped.sort_values(by = ('star_rating', 'mean'), ascending = False)
```
